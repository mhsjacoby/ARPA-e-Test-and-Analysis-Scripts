```python
import numpy as np
import matplotlib.pyplot as plt

import os
from datetime import datetime
```

```python
def mylistdir(directory):
    filelist = os.listdir(directory)
    return [x for x in filelist if x.endswith('.txt')]
```

```python
#path_filt = '/Users/maggie/Documents/Maggie-Grad-School/HPDmobile/USRP-Related/filter_test/'
path = '/Users/maggie/Documents/Maggie-Grad-School/HPDmobile/USRP-Related/USRP_data_collected/2019-10-21/1600/'

filtered_files = [x for x in os.listdir(path) if x.endswith('filtered.txt')]
raw_files = [x for x in os.listdir(path) if not 'filtered' in x]


def read_files(file_list):
    file_data = []
    print(len(file_list))
    for f in file_list:
        file1 = open(os.path.join(path, f), 'r') 
        new_data = file1.readlines()
        data = [float(x.strip('/n')) for x in new_data]
        file1.close()
        file_data = file_data + data
    return file_data
    

filtered = read_files(filtered_files)
raw = read_files(raw_files)

plt.figure(figsize = (15,5))
plt.plot(filtered)
plt.show()

plt.figure(figsize = (15,5))
plt.plot(raw)
plt.show()

# for f in sorted(filtered_files):
#     print(f)
# print(len(filtered_files))


# for f in sorted(raw_files):
#     print(f)
# print(len(raw_files))

# # filtered_files = mylistdir(path_filt)
# filtered_data = []


# for f in filtered_files:
#     file1 = open(os.path.join(path_filt,f), 'r') 
#     new_data = file1.readlines()
#     data = [float(x.strip('/n')) for x in new_data]
#     file1.close()
#     filtered_data = filtered_data + data
#     print(len(data))

# print(len(filtered_data))


```
