```python
# import os
# import sys
import csv
# import ast
# import json
from datetime import datetime
from datetime import timedelta
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib.pyplot import figure
import matplotlib.dates as mdates
plt.rcParams.update({'font.size': 16})
```

```python
D = 21

def plot_occ_asPD(occ_csv):
    df = pd.read_csv(occ_csv, delimiter = ',', index_col = 0, header = 0)
    df['Date'] = pd.to_datetime(df.index)         
    df = df.set_index('Date')
    dfs = []
    day1 = df.index[0]
    #for x in range()

    num_weeks = int(np.ceil(len(df)/(1440*D)))
    print('{} time periods of {} days'.format(num_weeks, D))
    for i in range(num_weeks):
        #weeks = D*i
        dayf = day1 + timedelta(days = D*i)
        dayn = day1 + timedelta(days = D*(i+1))
        print(dayf, dayn)
        dfn = df.loc[(df.index >= dayf) & (df.index < dayn)]
        dfs.append(dfn)
#     print(type(dfs), len(dfs))
    return dfs
```
